```python
# Remove Duplicates:
print(df.drop_duplicates())
```

```python
#  remove  Missing Data:
print(df.dropna())
```

```python
# Impute missing values
print(df.fillna(0))  # Replace missing values with 0
```

```python
# Resolve Inconsistencies

# Import the Scaler:
from sklearn.preprocessing import MinMaxScaler

# Normalize Specific Columns:
scaler = MinMaxScaler()
df[['total_rooms', 'median_income']] = scaler.fit_transform(df[['total_rooms', 'median_income']])

# Result
print(df.head())

# Replace Negative Values:
df['population'] = df['population'].apply(lambda x: 0 if x < 0 else x)

# Replace Values Outside an Expected Range:
df['latitude'] = df['latitude'].apply(lambda x: 90 if x > 90 else (-90 if x < -90 else x))
```

```python
# Steps to Remove Outliers Using IQR:

# Identify outliers for 'median_income' column
Q1 = df['median_income'].quantile(0.25)
Q3 = df['median_income'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df[(df['median_income'] >= lower_bound) & (df['median_income'] <= upper_bound)]

# Verify the result
print(df_no_outliers.shape)
```

```python
import matplotlib.pyplot as plt

# List of features to plot against the target variable
features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms',
            'total_bedrooms', 'population', 'households', 'median_income']

# Create subplots: 4 rows x 2 columns
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
fig.suptitle('Scatter Plots: Features vs Median House Value', fontsize=16)

# Flatten the axes array for easier iteration
axes = axes.ravel()

for i, feature in enumerate(features):
    axes[i].scatter(df[feature], df['median_house_value'], alpha=0.5)
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Median House Value')
    axes[i].set_title(f'{feature} vs Median House Value')
    axes[i].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

 ### Why Scatter Plots?
 Scatter plots allow us to:
1. Identify potential correlations or relationships between each feature and the target variable.
2. Detect patterns such as linear trends or clusters.
3. Spot any potential outliers.

```python
#Additional EDA Visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Exclude categorical columns to compute the correlation matrix
df_numeric = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = df_numeric.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Why a Heatmap?
It visually represents the strength and direction of correlations between numerical features.
Helps identify features that are strongly correlated with the target variable (median_house_value).
Useful for detecting multicollinearity between predictors, which can affect model performance.
